This notebook presents a detailed analysis of the Air Quality Index (AQI) in Delhi. The AQI is a measure used to communicate how polluted the air currently is or how polluted it is forecast to become. As air quality affects health, understanding its patterns is crucial, especially in urban areas like Delhi, known for significant air quality issues.

```python
# Focusing on Delhi data
delhi_df = city_day_df[city_day_df['City'] == 'Delhi']

# Handling missing values using forward fill
delhi_df = delhi_df.ffill().dropna()

# Converting date to datetime format for time series analysis
delhi_df['Date'] = pd.to_datetime(delhi_df['Date'])

# Set the date as the index (helpful for time-series analysis)
delhi_df.set_index('Date', inplace=True)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Define Diwali dates and assign a unique color to each date
diwali_dates = pd.to_datetime(['2015-11-11', '2016-10-30', '2017-10-19', '2018-11-07', '2019-10-27'])
colors = ['red', 'green', 'blue', 'purple', 'orange']
date_color_mapping = dict(zip(diwali_dates, colors))

# Full AQI data as background
plt.figure(figsize=(12, 6))
plt.scatter(delhi_df.index, delhi_df['AQI'], color='lightgray', label='Daily AQI')

# Highlighting Diwali dates with different colors
for date, color in date_color_mapping.items():
    diwali_aqi = delhi_df.loc[date, 'AQI']
    plt.scatter(date, diwali_aqi, color=color, zorder=5)

    # Highlighting the highest and lowest AQI in the Diwali month
    month_data = delhi_df[delhi_df.index.month == date.month]
    highest_aqi = month_data['AQI'].max()
    lowest_aqi = month_data['AQI'].min()

    highest_date = month_data[month_data['AQI'] == highest_aqi].index[0]
    lowest_date = month_data[month_data['AQI'] == lowest_aqi].index[0]

    plt.scatter(highest_date, highest_aqi, color='black', marker='^', zorder=5)  # Black triangle for highest
    plt.scatter(lowest_date, lowest_aqi, color='black', marker='v', zorder=5)  # Black inverted triangle for lowest

# Creating custom legend
legend_patches = [mpatches.Patch(color=color, label=date.strftime('%Y-%m-%d')) for date, color in date_color_mapping.items()]
legend_patches.append(mpatches.Patch(color='black', label='Highest/Lowest AQI'))
plt.legend(handles=legend_patches, title="Diwali Dates and AQI Extremes", bbox_to_anchor=(1.05, 1), loc='upper left')

# Adding labels and title
plt.title('AQI in Delhi with Diwali Days and AQI Extremes Highlighted')
plt.xlabel('Date')
plt.ylabel('AQI')
plt.tight_layout()
plt.show()
```

```python
import plotly.graph_objs as go
import plotly.express as px

# Define Diwali dates
diwali_dates = pd.to_datetime(['2015-11-11', '2016-10-30', '2017-10-19', '2018-11-07', '2019-10-27'])
colors = px.colors.qualitative.Set1

# Create a figure
fig = go.Figure()

# Plot the full AQI data
fig.add_trace(go.Scatter(x=delhi_df.index, y=delhi_df['AQI'], mode='markers', name='Daily AQI', marker=dict(color='lightgray', size=5)))

# Highlight Diwali dates with different colors and high/low AQI points
for i, date in enumerate(diwali_dates):
    diwali_aqi = delhi_df.loc[date, 'AQI']
    month_data = delhi_df[delhi_df.index.month == date.month]

    highest_aqi = month_data['AQI'].max()
    lowest_aqi = month_data['AQI'].min()

    highest_date = month_data[month_data['AQI'] == highest_aqi].index[0]
    lowest_date = month_data[month_data['AQI'] == lowest_aqi].index[0]

    # Adding Diwali, highest, and lowest AQI points
    fig.add_trace(go.Scatter(x=[date], y=[diwali_aqi], mode='markers', name=f'Diwali {date.year}', marker=dict(color=colors[i], size=10)))
    fig.add_trace(go.Scatter(x=[highest_date], y=[highest_aqi], mode='markers', name=f'Highest AQI {date.year}', marker=dict(symbol='triangle-up', color='black', size=12)))
    fig.add_trace(go.Scatter(x=[lowest_date], y=[lowest_aqi], mode='markers', name=f'Lowest AQI {date.year}', marker=dict(symbol='triangle-down', color='black', size=12)))

# Update layout
fig.update_layout(title='Interactive AQI Analysis in Delhi with Highlighted Diwali Days',
                  xaxis_title='Date',
                  yaxis_title='AQI',
                  hovermode='closest')

fig.show()
```
